```python
from math import exp, cos, sin, log


def f1(x):
    return exp(2 * x) * cos(3 * x)


def f2(x):
    return sin(log(x))


def f3(x):
    return log(x)


def f4(x):
    return cos(x) + sin(x)


def right_method(f, x, step):
    print(f"{right_method.__name__}, {f.__name__}, x0 = {x}, differentiation step = {step}: f'(x0) = ",
          (f(x + step) - f(x)) / step)


def left_method(f, x, step):
    print(f"{left_method.__name__}, {f.__name__}, x0 = {x}, differentiation step = {step}: f'(x0) = ",
          (f(x) - f(x - step)) / step)


def two_way_method(f, x, step):
    print(f"{two_way_method.__name__}, {f.__name__}, x0 = {x}, differentiation step = {step}: f'(x0) = ",
          (f(x + step) - f(x - step)) / (2 * step))


def second_order(f, x, step):
    return (f(x + 2 * step) - 2 * f(x) + f(x - 2 * step)) / ((2 * step) ** 2)


def third_order(f, x, step):
    return (f(x + 2 * step) - 2 * f(x + step) + 2 * f(x - step) - f(x - 2 * step)) / (2 * (step) ** 3)


f = [f1, f2, f3, f4]

x1 = [0.25, 0.51]
x2 = [0.99, 1.09]
x3 = [1.89, 2.39]

step = 0.0001

methods = [right_method, left_method, two_way_method]

print("\nFirst order derivative: ")

for x in x1:
    for method in methods:
        print("=======================================================================================================")
        for fun in f:
            method(fun, x, step)

print("\nSecond order derivative: ")

for x in x2:
    print("===================================================================================================")
    for fun in f:
        print(f"{fun.__name__}, x0 = {x}, differentiation step = {step}: f''({x}) = ", second_order(fun, x, step))

print("\nThird order derivative: ")

for x in x3:
    print("===================================================================================================")
    for fun in f:
        print(f"{fun.__name__}, x0 = {x}, differentiation step = {step}: f'''({x}) = ", third_order(fun, x, step))
```

```python
import pandas
import numpy as np
from scipy import interpolate

file = pandas.read_csv('owid-covid-data.csv')
data = file.iloc[21805:21962]

y_arr = data['new_cases']
y_arr = y_arr.values.tolist()

x_arr = [i for i in range(len(y_arr))]


def splines_method(x_0, x, y):
    tck = interpolate.splrep(x, y)

    return interpolate.splev(x_0, tck)


def squares_method(x, y):
    firstSum = 0
    secondSum = 0

    for i in range(0, len(x)):
        firstSum += x[i] * y[i]
        secondSum += x[i] ** 2

    first = (firstSum * len(x) - sum(x) * sum(y)) / (secondSum * len(x) - sum(x) ** 2)
    second = (np.sum(y) - first * sum(x)) / len(x)

    return first, second


def rectangle_method(x, y):
    res = 0

    for i in range(1, len(x)):
        h = x[i] - x[i - 1]
        res += y[int(x[i] - h / 2)]

    return res


first, second = squares_method(x_arr, y_arr)

func = [first * x + second for x in x_arr]

x_0 = np.linspace(0, 155, 155)

squares_rectangle = rectangle_method(x_arr, func)
spline_rectangle = rectangle_method(x_0, splines_method(x_0, x_arr, y_arr))

print(f"Least squares method + rectangle: {squares_rectangle}, error: {abs(squares_rectangle - 6434600)}")
print(f"Cubic spline method + rectangle: {spline_rectangle}, error: {abs(spline_rectangle - 6434600)}")
```
